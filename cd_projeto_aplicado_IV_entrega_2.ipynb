O problema identificado está relacionado à necessidade de atualização anual das tabelas de seguro empresarial. Segundo Morettin e Toloi (2018), a decomposição e modelagem de séries temporais permitem captar padrões cíclicos e sazonais essenciais para a tomada de decisão em contextos que envolvem dados históricos sequenciais.

# **2. Referencial Teórico**

A análise de séries temporais é uma das abordagens mais consolidadas para a compreensão e previsão de fenômenos que evoluem ao longo do tempo. Segundo Morettin e Toloi (2018), uma série temporal pode ser decomposta em componentes de tendência, sazonalidade e resíduo, permitindo identificar padrões estruturais que subsidiam a tomada de decisão em diversos setores. No contexto de acidentes de trabalho, essa decomposição é particularmente relevante, pois permite detectar ciclos sazonais associados a períodos de maior atividade econômica e variações regionais nos padrões de sinistros.

Entre os modelos estatísticos clássicos, o SARIMA (*Seasonal Autoregressive Integrated Moving Average*) é amplamente utilizado para séries com componente sazonal. Proposto como extensão do ARIMA por Box *et al.* (2015), o modelo incorpora parâmetros sazonais adicionais (P, D, Q, s) que capturam padrões repetitivos em intervalos regulares. Para dados mensais de acidentes de trabalho, espera-se sazonalidade com período s=12, o que torna o SARIMA uma escolha adequada como modelo de referência (*baseline*).

No campo do aprendizado de máquina, abordagens como o XGBoost têm sido aplicadas à previsão de séries temporais mediante a transformação do problema em regressão supervisionada por meio de *features* baseadas em *lags*, janelas móveis e variáveis temporais (JAMES *et al.*, 2013). Essa estratégia permite que modelos originalmente projetados para dados tabulares capturem dependências temporais, embora exija cuidados adicionais com a estacionariedade e a engenharia de atributos.

Outra abordagem relevante é o Prophet, desenvolvido por Taylor e Letham (2018) no Meta (Facebook). Trata-se de um modelo aditivo que decompõe a série em tendência, sazonalidade e efeitos de feriados, sendo robusto a dados faltantes e mudanças abruptas de tendência. Segundo Albeladi, Zafar e Mueen (2024), em estudo comparativo entre SARIMA, Prophet e Neural Prophet, cada modelo apresentou vantagens em cenários específicos: o SARIMA mostrou-se mais confiável para dados com sazonalidade bem definida, enquanto o Prophet se destacou em conjuntos menores com tendências claras.

No Brasil, a aplicação de séries temporais em dados previdenciários e de saúde ocupacional ainda é pouco explorada na literatura, embora os dados estejam disponíveis publicamente. O Observatório de Segurança e Saúde no Trabalho (SmartLab), mantido pelo Ministério Público do Trabalho em parceria com a OIT, disponibiliza painéis com dados agregados de acidentes e afastamentos (BRASIL, [s.d.]c), mas sem modelos preditivos integrados. Dessa forma, o presente projeto contribui ao aplicar técnicas modernas de previsão a um domínio de relevância social e econômica.

A combinação de abordagens estatísticas clássicas e de aprendizado de máquina permite uma avaliação comparativa robusta, utilizando métricas como MAE (*Mean Absolute Error*), RMSE (*Root Mean Squared Error*) e MAPE (*Mean Absolute Percentage Error*). Conforme Hyndman e Athanasopoulos (2021), a comparação de modelos heterogêneos por meio dessas métricas é fundamental para a seleção do modelo mais adequado ao contexto do problema.

# **3. Pipeline da Solução**

O pipeline proposto para o desenvolvimento do projeto está organizado em seis etapas principais, descritas a seguir.

**Etapa 1 – Coleta e Ingestão dos Dados.** Os dados serão obtidos diretamente do Portal de Dados Abertos do Governo Federal (dados.gov.br) e do Anuário Estatístico da Previdência Social (AEPS). Os arquivos em formato CSV serão carregados em ambiente Python (Google Colab) utilizando a biblioteca Pandas. Nesta etapa, serão verificadas a integridade dos arquivos, a consistência dos períodos e a completude das variáveis.

**Etapa 2 – Pré-processamento e Limpeza.** Os dados passarão por tratamentos de qualidade, incluindo: remoção ou imputação de valores ausentes, padronização de tipos de dados (datas, categorias), filtragem do período de interesse (2019–2024) e agregação mensal da variável-alvo (número total de acidentes por mês). Serão também criadas variáveis derivadas, como indicadores de região geográfica e setor econômico (CNAE).

**Etapa 3 – Análise Exploratória de Dados (EDA).** Serão realizadas análises descritivas e visuais para compreender a distribuição temporal dos acidentes, identificar padrões sazonais, tendências de crescimento e possíveis anomalias. As técnicas incluem: gráficos de série temporal, decomposição STL (*Seasonal and Trend decomposition using Loess*), autocorrelação (ACF) e autocorrelação parcial (PACF), além de análises segmentadas por UF e setor econômico.

**Etapa 4 – Modelagem e Treinamento.** Serão implementados e treinados três modelos de previsão: (i) SARIMA, como modelo estatístico de referência, com seleção de parâmetros via critério AIC/BIC; (ii) XGBoost, com engenharia de *features* temporais (*lags*, médias móveis, variáveis dummy de mês e ano); e (iii) Prophet, com configuração de sazonalidade anual e possíveis ajustes de feriados nacionais. A divisão dos dados seguirá a abordagem temporal: os últimos 12 meses serão reservados como conjunto de teste.

**Etapa 5 – Avaliação e Comparação dos Modelos.** Os modelos serão avaliados por meio das métricas MAE, RMSE e MAPE, comparando o desempenho preditivo no período de teste. Adicionalmente, será aplicada validação cruzada temporal (*time series cross-validation*) para avaliar a robustez das previsões. O modelo com melhor desempenho global será selecionado para a geração das previsões finais.

**Etapa 6 – Geração de Insights e Recomendações.** Com base nas previsões do modelo selecionado, serão geradas projeções de acidentes para os próximos 12 meses, segmentadas por região e setor econômico. Essas projeções serão traduzidas em recomendações estratégicas para ajuste das tabelas de seguro da KNRA Seguros, incluindo sugestões de regiões e setores com maior risco projetado.
